# Matrices

## Create matrices

```nim
import science/Matrix

let m = Matrix.createMatrix(@[@[1.0,1.0,1.0], @[2.0,2.0,2.0], @[3.0,3.0,3.0]])
echo(repr(m))
```

We can also create an empty (zero-ed) matrix by using `zeroMatrix` with the provided dimensions

```nim
import science/Matrix

let z = Matrix.zeroMatrix(3, 3)
echo(repr(z))
```

We can fill a matrix with random values by using the `randomMatrix` method:

```nim
#>flags -d:py3_version:3.6 -d:py3_dynamic --hints:off --verbosity:0
#>inlineplot 600 250
import science/Matrix

let z = Matrix.randomMatrix(13, 13)

show:
    imgshow(z.elements())
```

Matrix algebra can be performed using `+` for addition, `-` for subtraction and `*` for multiplication:

```nim
import science/Matrix

let z = Matrix.randomMatrix(3, 3)
echo(repr(z))
let zAdd = z + z
echo(repr(zAdd))
let zSub = z - z
echo(repr(zSub))
let zMult = z * z
echo(repr(zMult))
```

To transpose a matrix call the `transpose` method on the matrix

```nim
#>flags -d:py3_version:3.6 -d:py3_dynamic --hints:off --verbosity:0
#>inlineplot 600 250
import science/Matrix

let z = Matrix.randomMatrix(42, 42)
let zt = z.transpose()

show:
    imgshow(zt.elements())
```

Some math functions have shortcuts when applied to matrices, such as `abs`:

```nim
import science/Matrix

let m = Matrix.createMatrix(@[@[-1.0,1.0,-1.0], @[2.0,-2.0,2.0], @[3.0,3.0,-3.0]])
let a = m.abs()
echo(repr(a))
```

For methods not provided, the `map` method can be used:

```nim
import science/Matrix
from math import pow

let m = Matrix.createMatrix(@[@[-1.0,1.0,-1.0], @[2.0,-2.0,2.0], @[3.0,3.0,-3.0]])
let a = m.map(proc(x: float, i: int, j: int): float = pow(x, 2.0))
echo(repr(a))
```
